```python
#Loading packages

import numpy as np
import pylab as pl
%pylab inline 
```

```python
#Creating dictionary

md = {}
distributions = ['Normal', 'Poisson', 'Binomial', 'Chi-Square', 'Laplace', 'Gamma']

for d in distributions:
    md[d] = {}
```

```python
#Generating samples and including them as values in the dictionary

mymean = 100

mysize = np.random.randint(10, 2001, size=100)

for n in mysize:
    md['Normal'][n] = np.random.normal(mymean, size=n)
    md['Poisson'][n] = np.random.poisson(mymean, size=n)
    md['Binomial'][n] = np.random.binomial(2*mymean, 0.5, size=n)
    md['Chi-Square'][n] = np.random.chisquare(mymean, size=n)
    md['Laplace'][n] = np.random.laplace(mymean, size=n)
    md['Gamma'][n] = np.random.standard_gamma(mymean, size=n)
```

```python
# Creating a dictionary for the means of each distribution and populating it

means = {}

for d in distributions:
    means[d] = {}

for d in means:
    for s in md[d]:
        means[d][s] = md[d][s].mean()
```

```python
#Plotting all means

plt.figure(figsize=(10,10))
colors = list("rgbcmy")

for d in means.values():
    x = d.keys()
    y = d.values()
    plt.scatter(x,y,color=colors.pop())
    plt.xlim(0, 2000)
    plt.ylim(min(y)-0.5, max(y)+0.5)
    plt.legend(means.keys())
    
hlines(100,0,2000)
plt.show()
```

ANSWER: As the size of the samples increase, the means seem to convert towards the population mean of 100.
This expected according to the Central Limit Theorem (i.e. the distribuion of the means will approach a normal distribution).

```python
#Creating list with all means

allmeans = []

for d in means:
    for s in md[d]:
        allmeans.append(means[d][s])
```

```python
#Plotting histogram

plt.figure(figsize=(10,10))
plt.hist(allmeans, bins=30)
plt.xlabel("Sample mean")
plt.ylabel("N")
plt.show()
```
